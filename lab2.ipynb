```python
data['Is_Excellent']=data['Grade'].apply(lambda x: 'True' if x>=90 else 'False')
data
```

```python
data['Grade']=data['Grade'].fillna(data['Grade'].mean())
data
```

```python
data['Score Date']=data['Score Date'].fillna(data['Score Date'].mode()[0])
data['Score Date']
```

```python
data['Final_Grade']=data['Grade']+5
data
```

```python
data['Grade_Category']=pd.cut(data['Grade'],bins=[0,69,79,89,100],labels=['Fail','Average','Good','Excellent'],
                    include_lowest=True)
data
```

```python
data=data.drop(columns='Student ID')
data
```

```python
data=data[data['Passed']==True]
data
```

```python
data=data.sort_values('Grade',ascending=False)
data
```

```python
data=data.sort_values(['City','Grade'],ascending=[True,False])
data
```

```python
data['Rank']=data['Grade'].rank(ascending=False,method='min')
data
```

```python
max_grades=data['Grade'].max()
highest_students=data[data['Grade']==max_grades][['Name','Grade']]
highest_students
```
