```python
#9
mat=np.array([
    [4,8,1,12],
    [7,3,9,5],
    [11,2,6,10],
    [15,0,14,13]
]
)
#1
print(mat.mean(axis=1,keepdims=True))
#2
print(mat.sum(axis=0,keepdims=True))
#3
print("Max value",mat.max())
pos=np.unravel_index(mat.argmax(), mat.shape)
print("Position:",pos)

```
