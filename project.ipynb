```python
def parse_address(addr_str):
    try:
        addr_dict = json.loads(addr_str.replace("'", "\""))  # fix single quotes to double
        return addr_dict.get('city', 'Unknown') 
    except:
        return 'Unknown'

df['city'] = df['address'].apply(parse_address)
```

```python
df.head(5)
```

```python
df['city'].value_counts()
```

```python
def parse_address2(addr_str):
    try:
        addr_dict = json.loads(addr_str.replace("'", "\"")) 
        return addr_dict.get('country', 'Unknown') 
    except:
        return 'Unknown'

df['country'] = df['address'].apply(parse_address2)
```

```python
df['country'].value_counts()
```

```python
def parse_address3(addr_str):
    try:
        addr_dict = json.loads(addr_str.replace("'", "\"")) 
        return addr_dict.get('state', 'Unknown') 
    except:
        return 'Unknown'

df['state'] = df['address'].apply(parse_address3)
```

```python
df['state'].value_counts()
```

```python
df['role'].value_counts()
```

```python
def parse_bank(bank_str):
    try:
        bank_dict = json.loads(bank_str.replace("'", "\"")) 
        return bank_dict.get('cardType', 'Unknown') 
    except:
        return 'Unknown'
    
df['cardType']=df['bank'].apply(parse_bank)
```

```python
print('nuniques',df['cardType'].nunique())
print(df['cardType'].value_counts())
```

```python
def parse_bank2(bank_str):
    try:
        bank_dict = json.loads(bank_str.replace("'", "\"")) 
        return bank_dict.get('currency', 'Unknown') 
    except:
        return 'Unknown'
    
df['currency']=df['bank'].apply(parse_bank2)
```

```python
print('nuniques',df['currency'].nunique())
print(df['currency'].value_counts())
```

```python
def parse_hair(hair_str):
    try:
        hair_dict = json.loads(hair_str.replace("'", "\"")) 
        return hair_dict.get('type', 'Unknown') 
    except:
        return 'Unknown'
    
df['hair_type']=df['hair'].apply(parse_hair)
```

```python
print('nuniques',df['hair_type'].nunique())
print(df['hair_type'].value_counts())
```

```python
def parse_company(company_str):
    try:
        company_dict = json.loads(company_str.replace("'", "\"")) 
        return company_dict.get('title', 'Unknown') 
    except:
        return 'Unknown'
    
df['job_title']=df['company'].apply(parse_company)
```

```python
print('nuniques',df['job_title'].nunique())
print(df['job_title'].value_counts())
```

```python
average_age = df['age'].mean()
print(f"Average age of users: {average_age:.2f}")
```

```python
sns.histplot(df['age'],
            color='teal',
            kde=True,
            edgecolor='black',
            alpha=0.8)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python

data=df.groupby('gender')['age'].mean()
plt.figure(figsize=(5, 5))
plt.pie(data, 
        labels=['Female', 'Male'], 
        autopct='%1.1f%%', 
        colors=['#66b3ff', '#ff9999'],
        startangle=90)
plt.title('Average Age by Gender', fontsize=14)
plt.axis('equal')  
plt.show()
```

```python
counts=df['gender'].value_counts()
plt.figure(figsize=(5,4))
plt.bar(counts.index,counts.values,color='teal', edgecolor='black')
plt.title('Number of Users per Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
counts=df['cardType'].value_counts()
plt.figure(figsize=(8,4))
plt.bar(counts.index,counts.values,color='teal', edgecolor='black')
plt.title('Number of Users per cardtype', fontsize=14)
plt.xlabel('CardType', fontsize=10)
plt.ylabel('Number of Users', fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45) 
plt.show()
```

```python
country_counts = df.groupby('city').size().reset_index(name='User_Count')

country_counts = country_counts.sort_values('User_Count', ascending=False)

print(country_counts.head(10))
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=country_counts.head(10),
            x='city',          
            y='User_Count', 
            palette='viridis')
plt.title('Number of Users by Country (Top 10)', fontsize=15)
plt.ylabel('Number of Users', fontsize=10)
plt.xlabel('Country', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
